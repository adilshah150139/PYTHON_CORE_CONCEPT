💼 Different Job Salaries & How to Deal with None in Python 🐍


```python
job_titles = np.array(['Data Analyst', 'Data Scientist', 'Data Engineer', 'Machine Learning Engineer', 'AI'])

# Base salaries
base_salaries = np.array([4449, 3000, 6000, 1200, 4000])

# when use 0
bonus_rates_include_0 = np.array([0.2, 0.1, 0.3, .1, 0])

# Bonus rates
bonus_rates_nan = np.array([0.2, 0.1, 0.3, .1, np.nan])  # Nane (no bonus),because 0 effect calculation


```

```python
total_salaries_0 = base_salaries * (1 + bonus_rates_include_0)
total_salaries_nan = base_salaries * (1 + bonus_rates_nan)
print(total_salaries_0)
print("Mean Salaries including 0: ", np.mean(total_salaries_0))
print("Mean Salaries using nan: ", np.mean(total_salaries_nan))
```

```python
# using nan keyword from numpy
print(np.nanmean(total_salaries_nan))
```
