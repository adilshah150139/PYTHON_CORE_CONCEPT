### 📊 Pandas Aggregation Methods
We will cover all these methods
| Method | Text Version | Description |
|--------|--------------|-------------|
| `df.count()` | `'count'` | Count non-null values |
| `df.sum()` | `'sum'` | Sum of values |
| `df.min()` | `'min'` | Minimum value |
| `df.idxmin()` | `'idxmin'` | Index of minimum value |
| `df.max()` | `'max'` | Maximum value |
| `df.idxmax()` | `'idxmax'` | Index of maximum value |
| `df.mean()` | `'mean'` | Arithmetic mean/average |
| `df.median()` | `'median'` | Median value |
| `df.std()` | `'std'` | Standard deviation |
| `df.var()` | `'var'` | Variance |
| `df.quantile()` | `'quantile'` | Quantile/percentile |
| `df.describe()` | `'describe'` | Full statistical summary |

```python
import pandas as pd
from datasets import load_dataset

# Load datasets
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])  # convert from str to int mean date stamp
print(type(df['job_posted_date'][0]))
```

```python
df.describe()   # describe() method only show Numeric columns
```

```python
df['salary_year_avg'].median()
```

```python
min_salary = df['salary_year_avg'].idxmin() # Find index of Minimum salary
df.iloc[min_salary]
```

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median() # or min() and max() as well
```

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min','max', 'median']) # finding as combined form
```
