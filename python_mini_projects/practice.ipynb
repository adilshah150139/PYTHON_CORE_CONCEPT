```python
def get_grade_letter(scores):
    if scores >= 90:
        return 'A'
    elif scores >= 80:
        return 'B'
    elif scores >= 70:
        return 'C'
    elif scores >= 60:
        return 'E'
    else:
        return 'F'
    
```

```python

def calculate_score(scores):
    average = sum(scores)/ len(scores)
    highest = max(scores)
    lowest = min(scores)
    return round(average, 2), highest, lowest
```

```python

def display_report(name, subjects, scores):
    print("=" * 40)
    print(f"GRADE REPORT FOR: {name.upper()}")
    print("=" * 40)

    for subject, score in zip(subjects, scores):
        letter = get_grade_letter(score)
        print(f"{subject:<20}  {score:>3} ({letter})")

    """  Formatting Notes:
    - {subject:<20} : Left-align subject name in 20 spaces (creates clean columns)
    - {score:>3}    : Right-align score in 3 spaces (numbers line up nicely)
    - {letter}      : Grade letter from get_grade_letter()
    """

    print("-" * 40)
    avg, high, low = calculate_score(scores)
    print(f"  Average: {avg}")
    print(f"  Highest: {high}")
    print(f"  Lowest:  {low}")
    print(f"  Over all-Grades:  {get_grade_letter(avg)}")
    print("=" * 40)


   # --- Sample data ---
student_name = "Adil shah"
subjects = ["Calculus", "Linear Algebra", "Physics", "Discrete Math", "Chinese"]
scores = [82, 76, 88, 71, 95]
display_report(student_name, subjects, scores)
```

### To_do List:

```python
import os

FILENAME = "todo_list.txt"
def load_task():
    if not os.path.exists(FILENAME):
        return []
    with open(FILENAME, 'r') as f:
        return [line.strip() for line in f.readlines() if line.strip()]

# save_data
def save_task(tasks):
    with open(FILENAME, 'w') as f:
        for task in tasks:
            f.write(task + "\n")

# OutPut:
def show_task(tasks):
    if not tasks:
        return 
    print("\n ---Your To-Do LIST---")
    for i, task in enumerate(tasks, 1):
        print(f"  {i}. {task}")
    print(f"  Total: {len(tasks)} task(s)")

# append task:
def add_task(tasks, task):
    tasks.append(task)
    save_task(tasks)
    print(f"✓ Added: '{task}'")


# Remove tasks:
def remove_task(tasks, number):
    if 1 <= number <= len(tasks):
        removed = tasks.pop(number - 1)
        save_task(tasks)
        print(f"✓ Removed: {removed}")
    else:
        print("Invalid task number")
        

# Demo:
tasks = load_task()
add_task(tasks, "Study Calculus Chapter 9")
add_task(tasks, "Practice Python loops")
add_task(tasks, "Review Linear Algebra notes")
add_task(tasks, "Pray Fajr and read Quran")

show_task(tasks)
remove_task(tasks,5)


```
