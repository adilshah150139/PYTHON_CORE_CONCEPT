### What ast Does:
ast stands for Abstract Syntax Tree. It's a module that helps Python understand and work with Python code itself!

```python
import ast
string = "[1,2,3,4]"   
list = ast.literal_eval(string)   # convert a str to list
print(type(string))
print(type(list))
```
